**Stemming**

Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as a lemma. Stemming is important in natural language understanding (NLU) and natural language processing (NLP).

````python
words=['eating', 'eats', 'eaten', 'writing', 'writes', 'programming', 'programs', 'history', 'finally', 'finalized']
````

**PorterStemmer**

````python
from nltk.stem import PorterStemmer
````

````python
stemming=PorterStemmer()
````

````python
for word in words:
  print(f"{word} ----> {stemming.stem(word)}")
````

**RegexpStemmer class**

NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example

````python
from nltk.stem import RegexpStemmer
````

````python
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
````

````python
reg_stemmer.stem("eating")
````

````python
reg_stemmer.stem("writing")
````

### üîπ **Snowball Stemmer ‚Äì Short Definition:**

> **Snowball Stemmer** is a more advanced stemming algorithm provided by NLTK that removes common suffixes from words to reduce them to their root (stem) form.
> It supports **multiple languages** and produces more accurate stems than simpler stemmers like Porter.



### üîç Key Points:

* Supports **15+ languages** (like English, French, German, etc.)
* More consistent and aggressive than PorterStemmer
* Built using the **Snowball language** (that‚Äôs why the name)


---



````python
from nltk.stem import SnowballStemmer
````

````python
snowballsstemmer=SnowballStemmer('english')
````

````python
for word in words:
  print(f"{word} -----> {snowballsstemmer.stem(word)}")
````


## ü•ä **Porter Stemmer vs Snowball Stemmer**

| Feature                    | **Porter Stemmer**             | **Snowball Stemmer**                          |
| -------------------------- | ------------------------------ | --------------------------------------------- |
| üîç **Developer**           | Martin Porter                  | Martin Porter (enhanced version)              |
| üöÄ **Speed**               | Fast                           | Slightly slower (but negligible difference)   |
| üéØ **Accuracy**            | Decent                         | More accurate and consistent                  |
| üìö **Rules**               | Older, hardcoded rules         | Modular, extensible rule system               |
| üåç **Languages Supported** | English only                   | 15+ languages (English, French, German, etc.) |
| üì¶ **Library**             | `nltk.stem.PorterStemmer`      | `nltk.stem.SnowballStemmer`                   |
| ü§ñ **Stemming Style**      | Conservative (less aggressive) | More aggressive and consistent                |
| üîß **Customizability**     | Not customizable               | Somewhat customizable with language settings  |

**Wordnet Lemmatizer**

Lemmatization technique is like stemming. The output we will get after lemmatization is called 'lemma', which is a root word rather than root stem, the output of stemming. After lemmatization, we will be getting a valid word that means the same thing.

NLTK provides WordNetLemmatizer class which is a thin wrapper around the wordnet corpus. This class uses `morphy()` function to the WordNet CorpusReader class to find a lemma. Let us understand it with an example ‚Äì


## üìö **WordNet Lemmatizer ‚Äì Short Definition:**

> **WordNet Lemmatizer** is a text processing tool in NLTK that reduces a word to its base or dictionary form (**lemma**), using the meaning and part-of-speech.
> Unlike stemmers, it produces real words as output.

---

## ‚úÖ **Key points:**

* Uses the **WordNet lexical database** (a large English dictionary).
* Needs the **part of speech (POS)** to get better results (noun, verb, adjective, etc.).
* Returns actual English words (lemmas), unlike stemmers which may give truncated roots.

---

## üîß **Example:**

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("running", pos='v'))   # run
print(lemmatizer.lemmatize("better", pos='a'))    # good
print(lemmatizer.lemmatize("cats"))               # cat
```

### üß† Without POS:

```python
print(lemmatizer.lemmatize("running"))   # running (default is noun)
```

Default POS is noun (`'n'`), so sometimes you won‚Äôt get the expected lemma.

---

## üîç **Difference from Stemming:**

|                | Stemming             | Lemmatization                   |
| -------------- | -------------------- | ------------------------------- |
| Meaning-based? | ‚ùå No                 | ‚úÖ Yes (uses WordNet dictionary) |
| Output         | May not be real word | Always real word                |
| POS needed?    | ‚ùå No                 | ‚úÖ Better if POS is given        |

---

## ‚öô **How to use POS with lemmatizer:**

* `'n'` ‚Üí noun
* `'v'` ‚Üí verb
* `'a'` ‚Üí adjective
* `'r'` ‚Üí adverb

---




````python
nltk.download('wordnet')
````

````python
from nltk.stem import WordNetLemmatizer
````

````python
lemmatizer=WordNetLemmatizer()
````

````python
lemmatizer.lemmatize("going") # running (default is noun)
````

````python
lemmatizer.lemmatize("cat",pos='n') #'n' ‚Üí noun
````

````python
lemmatizer.lemmatize("going",pos='v') #'v' ‚Üí verb
````

````python
lemmatizer.lemmatize("better",pos='a') #'a' ‚Üí adjective
````

````python
lemmatizer.lemmatize("better",pos='r') #'r' ‚Üí adverb
````



## üîç **Examples & what happens:**

| Code                                      | POS                    | Meaning        | Lemma / Output        |
| ----------------------------------------- | ---------------------- | -------------- | --------------------- |
| `lemmatizer.lemmatize("going")`           | default (`'n'` = noun) | Noun form      | `'going'` (no change) |
| `lemmatizer.lemmatize("cat", pos='n')`    | noun                   | Cat stays same | `'cat'`               |
| `lemmatizer.lemmatize("going", pos='v')`  | verb                   | Verb form      | `'go'`                |
| `lemmatizer.lemmatize("better", pos='a')` | adjective              | Adjective form | `'good'`              |
| `lemmatizer.lemmatize("better", pos='r')` | adverb                 | Adverb form    | `'well'`              |

---



## ‚öô **Time Complexity Summary Table**

| Algorithm              | Typical Time Complexity (per word) | Notes                                                                                             |
| ---------------------- | ---------------------------------- | ------------------------------------------------------------------------------------------------- |
| **Porter Stemmer**     | `O(k)`                             | k = length of word; uses simple rule-based suffix stripping                                       |
| **Snowball Stemmer**   | `O(k)`                             | Similar to Porter, but rules are modular and language-aware                                       |
| **RegexpStemmer**      | `O(k¬∑p)`                           | p = number of regex patterns; applies regex sequentially                                          |
| **Lancaster Stemmer**  | `O(k)`                             | Rule-based, quite aggressive                                                                      |
| **WordNet Lemmatizer** | `O(k + log n)`                     | Looks up in WordNet dictionary of size n; search is fast but slower than pure rule-based stemmers |
